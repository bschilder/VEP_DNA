## 1000 Genomes Project

The 1000 Genomes Project (1KG) has a lot of different versions of their genomic data, as well as additional data that came from offshoot projects. These are referred to as *collections* and a description of each can be found [here](https://www.internationalgenome.org/data-portal/data-collection).

The `onekg` submodule provides a way to easily find and download data from those projects, namely VCF files.


We can list all the different collections currently supported by the `onekg` submodule with the `get_ftp_dict` function. 

This also returns information about the manifest file (which contains metadata about the names and expected sizes of each file) as well as info about how to read each manifest file into `pandas` (used internally by other functions).

### VCFs

```python
og.get_ftp_dict()
```

The `list_remote_vcf` function returns a manifest of the VCF files available for download for a given collection.

```python
manifest  = og.list_remote_vcf(key="1000_Genomes_30x_on_GRCh38")
manifest.head()
```

Now we can select a collection and download the VCF files with the `download_vcfs` function. In this case, we are selecting the `"1000_Genomes_30x_on_GRCh38"` collection, which contains VCFs of high-coverage WGS data from 3202 individuals.

**NOTE**: These files are quite large and can take a long time to download. 

```python
vcf_files = og.download_vcfs(key="1000_Genomes_30x_on_GRCh38")
```

We can also donwload VCFs from another collection, such as the `"Human_Genome_Diversity_Project"` collection, which contains VCFs of WGS data from 929 diverse individuals.

These VCFs files were not directly provided by 1KG, but instead post-processed by the Sanger Insitute and distributed on a separate server. This is all handled automatically by `download_vcfs`.

**NOTE**: These files are quite large and can take a long time to download. 

```python
vcf_files = og.download_vcfs(key="Human_Genome_Diversity_Project")
```

### Metadata

`onekg` has additional functions for getting relevant metadata like sample IDs and population (ancestry) IDs.

**Population metadata**: 
Contains information about each population, and which superpopulation they're part of.

```python
pop = og.get_pop()
pop.head()
```

**Pedigree metadata**: 
contains information about how some individuals within the cohort are related to each other.

```python
ped = og.get_ped()
ped.head()
```

`get_sample_metadata` is a function that merges together the pedigree and population metadata to give a mapping for each individual to their population and superpopulation.

```python
sample_metadata = og.get_sample_metadata()
sample_metadata.head()
```
