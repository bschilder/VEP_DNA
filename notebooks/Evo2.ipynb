# Evo 2

## Install

Create a conda env with some dependencies installed.

```
conda env create -f /grid/koo/home/schilder/projects/VEP_DNA/conda/evo2.yml
```

Activate that env:
```
conda activate evo2
```

```
pip3 install torch torchvision torchaudio
```

```
git clone --recurse-submodules git@github.com:ArcInstitute/evo2.git
cd evo2
```

Make sure the pip installation within your conda env is the one being used
```
export PATH="$HOME/miniconda3/envs/evo2/bin:$PATH"
```

Export the path to your the `cudnn` installation within your conda env.
NOTE: Make sure to change the `python3.11` to the actual version you're using (if you change the conda env)
```
export CUDNN_PATH=/path/to/cudnn
export CPATH=$HOME/miniconda3/envs/evo2/lib/python3.11/site-packages/nvidia/cudnn/include:$CPATH
```

```
module load gcc9
module load cuda12.3/toolkit/12.3.2
```


```
# python setup.py install
pip install .
```

Export the path to your the `cudnn` installation within your conda env.
```
export CPATH=$HOME/miniconda3/envs/evo2/lib/python3.13/site-packages/nvidia/cudnn/include:$CPATH
```
